This notebook analyzes and visualizes global COVID-19 data using Python.

## Objectives
- Load and clean COVID-19 dataset
- Explore global trends in confirmed cases, deaths, and recoveries
- Visualize data with charts
- Gain insights into the pandemic progression

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# For cleaner charts
sns.set(style="darkgrid")
```

## Load Dataset
We will use the COVID-19 dataset from Our World in Data (sample CSV link).

```python
# Load sample dataset from Our World in Data (CSV)
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
data = pd.read_csv(url)

# Show first 5 rows
data.head()
```

## Data Cleaning
Keep only important columns.

```python
# Keep relevant columns
df = data[["location", "date", "total_cases", "total_deaths", "total_vaccinations"]]

# Drop rows with missing values
df = df.dropna()

df.head()
```

## Global COVID-19 Cases Over Time

```python
# Filter for World data
world = df[df["location"] == "World"]

# Plot cases and deaths
plt.figure(figsize=(12,6))
plt.plot(world["date"], world["total_cases"], label="Total Cases")
plt.plot(world["date"], world["total_deaths"], label="Total Deaths")
plt.xticks(rotation=45)
plt.title("Global COVID-19 Cases & Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend()
plt.show()
```

## Top 10 Countries by Total Cases

```python
# Get latest date
latest_date = df["date"].max()
latest = df[df["date"] == latest_date]

# Top 10 countries by cases
top10 = latest.sort_values("total_cases", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="total_cases", y="location", data=top10, palette="Reds_r")
plt.title("Top 10 Countries by COVID-19 Cases")
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.show()
```
